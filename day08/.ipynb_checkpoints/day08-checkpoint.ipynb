# Part 1

```python
from itertools import combinations
from math import dist, prod

with open("input.txt") as f:
    nodes = [tuple(map(int, line.split(","))) for line in f]
    
distances = [(dist(nodes[x], nodes[y]), x, y) for x, y in combinations(range(len(nodes)), r=2)]
distances.sort()

sets = []
connections = 1000

for _, a, b in distances:
    found = [s for s in sets if a in s or b in s]
    for f in found:
        sets.remove(f)
    sets.append({a, b}.union(*found))
    connections -= 1

    if connections <= 0: break
    
sets.sort(key=len, reverse=True)
print(prod(len(s) for s in sets[:3]))
```

# Part 2

```python
from itertools import combinations
from math import dist

file = "input.txt"
with open(file) as f:
    nodes = [tuple(map(int, line.split(","))) for line in f]
    
distances = [(dist(nodes[x], nodes[y]), x, y) for x, y in combinations(range(len(nodes)), r=2)]
distances.sort()

sets = []

for _, a, b in distances:
    found = [s for s in sets if a in s or b in s]
    for f in found:
        sets.remove(f)
    sets.append({a, b}.union(*found))      

    if len(sets) == 1 and len(sets[0]) == len(nodes): break
        
print(nodes[a][0]*nodes[b][0])
```
