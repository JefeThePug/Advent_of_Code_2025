```python
import re

with open("input.txt") as f:
    range_data, ingredients = f.read().split("\n\n")
    
ranges = [range(int(a), int(b)+1) for a, b in re.findall(r"(\d+)-(\d+)", range_data)]

fresh = sum(any(n in r for r in ranges) for n in map(int, ingredients.split("\n")))
print(fresh)
```

```python
import re

with open("input.txt") as f:
    range_data, _ = f.read().split("\n\n")
    
ranges = sorted((int(a), int(b)) for a, b in re.findall(r"(\d+)-(\d+)", range_data))

merged = []
for start, end in ranges:
    if not merged or start > merged[-1][1] + 1:
        merged.append([start, end])
    else:
        merged[-1][1] = max(merged[-1][1], end)

print(sum(b - a + 1 for a, b in merged))
```
