# Part 1

```python
import numpy as np
from numpy.lib.stride_tricks import sliding_window_view

with open("input.txt") as f:
    grid = np.array([list(line) for line in f.read().split("\n")])
    grid = np.pad(grid, pad_width=1, mode="constant", constant_values=".")

windows = sliding_window_view(grid, (3, 3))
centers = windows[:, :, 1, 1] == "@"
counts = (windows == "@").sum(axis=(2, 3))
total = (centers & (counts < 5)).sum()

print(total)
```

```python
import numpy as np
from numpy.lib.stride_tricks import sliding_window_view

with open("input.txt") as f:
    grid = np.array([list(line) for line in f.read().split("\n")])
    grid = np.pad(grid, pad_width=1, mode="constant", constant_values=".")

total = 0

while True:
    windows = sliding_window_view(grid, (3, 3))
    centers = windows[:, :, 1, 1] == "@"
    counts = (windows == "@").sum(axis=(2, 3))
    
    remove_mask = centers & (counts < 5)
    if (removed:=remove_mask.sum()) == 0:
        break
    total += removed

    grid[1:-1, 1:-1][remove_mask] = "."

print(total)
```

```python
import numpy as np
from numpy.lib.stride_tricks import sliding_window_view

def scan(grid):
    windows = sliding_window_view(grid, (3, 3))
    centers = windows[:, :, 1, 1] == "@"
    counts = (windows == "@").sum(axis=(2, 3))
    remove_mask = centers & (counts < 5)
    return remove_mask.sum(), remove_mask

with open("input.txt") as f:
    grid = np.array([list(line) for line in f.read().split("\n")])
    grid = np.pad(grid, pad_width=1, mode="constant", constant_values=".")
    
# Part 1
total, _ = scan(grid)
print(total)

# Part 2
total = 0

while True:
    removed, remove_mask = scan(grid)
    if removed == 0:
        break
    total += removed
    grid[1:-1, 1:-1][remove_mask] = "."

print(total)
```
