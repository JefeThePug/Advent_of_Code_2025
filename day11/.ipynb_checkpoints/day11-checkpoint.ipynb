```python
import re
from functools import cache
from math import prod

def count_paths(graph, start, goal):
    @cache
    def dp(node):
        if node == goal:
            return 1
        return sum(dp(nxt) for nxt in graph.get(node, []))
    return dp(start)
    
    
with open("input.txt") as f:
    graph = {k: v.split() for k, v in re.findall(r"([a-z]{3}): ([ a-z]+)", f.read())}
    
    
paths = prod([
    count_paths(graph, "svr", "fft"),
    count_paths(graph, "fft", "dac"),
    count_paths(graph, "dac", "out")
])

print(paths)
```
