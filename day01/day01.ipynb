# PART 1

```python
point = 50
zeros = 0
dirs = {"L": -1, "R": 1}

with open("input.txt") as f:
    for line in f:
        point = (point + int(line[1:])*dirs[line[0]]) % 100
        zeros += (point == 0)
        
print(zeros)
```
