```python
from math import prod

def operate(nums, operator):
    return sum(nums) if operator == "+" else prod(nums)

with open("input.txt") as f:
    grid = ["".join(n).strip() for n in zip(*f)]
    
new_nums, ops = [], []

for line in grid:
    if line == "": continue
    if line.isdigit():
        new_nums[-1].append(int(line))
    else:
        ops.append(line[-1])
        new_nums.append([int(line[:-1].strip())])

print(sum(operate(n, o) for n, o in zip(new_nums, ops)))
```
