```python
import re
from math import log10

with open("input.txt") as f:
    ranges = [range(int(i), int(j)+1) for i, j in re.findall(r"(\d+)-(\d+)", f.read())]
    
total = 0

for r in ranges:
    for n in r:
        half = (int(log10(n)) + 1) // 2
        a, b = divmod(n, 10**half)
        if a == b:
            total += n

print(total)
```

```python
import re

with open("input.txt") as f:
    ranges = [range(int(i), int(j)+1) for i, j in re.findall(r"(\d+)-(\d+)", f.read())]
    
invalids = set()

for n in range(1, 8):
    for r in ranges:
        line = f',{",".join(f"{x}" for x in r)},'
        repeats = r"(\1)"*n
        pattern = fr"\D([1-9]\d*){repeats}\D"
        for i in re.findall(pattern, line):
            invalids.add(int("".join(i)))

print(sum(invalids))
```
