```python
import numpy as np

with open("input.txt") as f:
    grid = np.array([list(row) for row in f.read().split("\n")])

beams = {np.argwhere(grid == "S")[0][1]}
splits = set()

for r in range(grid.shape[0]):
    new_beams = set()
    for c in beams:
        if grid[r, c] == "^":
            splits.add((r, c))
            new_beams.update({c-1, c+1})
        else:
            new_beams.add(c)
    beams = new_beams

print(len(splits))
```

```python
import numpy as np
from functools import cache

@cache
def fall(r, c):
    if r + 1 == grid.shape[0]:
        return 1
    
    if grid[r+1, c] == ".":
        return fall(r+1, c)
    else:
        return fall(r+1, c-1) + fall(r+1, c+1)

with open("input.txt") as f:
    grid = np.array([list(row) for row in f.read().split("\n")])
    
fall(*map(int, np.where(grid == "S")))
```
