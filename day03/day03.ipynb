```python
def get_jolt(arr, depth):
    if depth == 0: 
        return 0
    
    for n in range(9, -1, -1):
        if f"{n}" not in arr: 
            continue
        tail = get_jolt(arr[arr.index(f"{n}")+1:], depth-1)
        if tail >= 0:
            return n*(10**(depth - 1)) + tail
    
    return -1


with open("input.txt") as f:
    lines = f.readlines()

# Part 1
print(sum(get_jolt(line, 2) for line in lines))

# Part 2
print(sum(get_jolt(line, 12) for line in lines))
```
